## What product are most often sold together

```python
#this means that when products have the same order ID they are sold together
all_months['Order ID'].value_counts()
```

```python
#The join() method takes all items in an iterable and joins them into one string.
#myTuple = ("John", "Peter", "Vicky")
#x = "#".join(myTuple)
# John#Peter#Vicky
#Input:
#apply implicitly passes all the columns for each group as a DataFrame to the custom function.
#while transform passes each column for each group individually as a Series to the custom function.
#So, transform works on just one Series at a time and apply works on the entire DataFrame at once.
def join(x):
    return ','.join(x)
all_months['Grouped']=all_months['Product'].groupby(all_months['Order ID']).transform(join)
all_months
```

```python
all_months[all_months['Order ID']=='235798']
```

```python
order=all_months[['Order ID','Grouped']]
order=order.drop_duplicates()
```

```python
order
```

```python
order['Grouped'].value_counts()
#order[order['Grouped']=='Google Phone,USB-C Charging Cable,Wired Headphones,27in FHD Monitor']
```

```python
#Pythonâ€™s Itertool is a module that provides various functions that work on iterators to produce complex iterators. 
from itertools import combinations
from collections import Counter
#A Counter is a subclass of dict.
```

```python
#txt = "hello, my name is Peter, I am 26 years old"
#x = txt.split(", ")
#['hello', 'my name is Peter', 'I am 26 years old']
count=Counter()
for row in order['Grouped']:
    row_list=row.split(",")
    #count.update(data)
    count.update(Counter(combinations(row_list,3)))
count.most_common(10)
```

## What product sold the most? And why

```python
product_ordered=all_months.groupby(['Product']).sum()
product_ordered
```

```python
plt.bar(product_ordered.index.values,product_ordered['Quantity Ordered'],color='red')
plt.xlabel("Products")
plt.xticks(rotation='vertical')
plt.ylabel("Quantity")
```

```python
#the quantity sold depends on the price , higher the price lower the same so lets do a price analysis
price=all_months.groupby(['Product']).mean()
price
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(product_ordered.index.values,product_ordered['Quantity Ordered'],color='r')
ax2.plot(product_ordered.index.values,price['Price Each'],'b-')

ax1.set_xlabel('Products')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price', color='b')
ax1.set_xticklabels(labels=product_ordered.index.values,rotation='vertical')
plt.show()
```
